```python
with open('my_friends.html') as file:
    text = file.read()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(text, 'lxml')
friends_list = soup.findAll('div', {'class': 'friends_user_row'})

friends = []

for item in friends_list:
    friends.append(item.find('div', {'class': 'friends_field_title'}).find('a').text)
```

```python
for i, friend in enumerate(friends):
    n, s, *_ = friend.split()
    s, n = n, s
    friends[i] = n + ' ' + s
friends.append('Мурашов Леонид')
friends
```

```python
import urllib.request

admlist_adress ='http://admlist.ru'
with urllib.request.urlopen(admlist_adress + '/index.html') as file:
    text = file.read().decode('utf-8')

soup = BeautifulSoup(text, 'lxml')
VUZ_list = soup.findAll('tr')

links = []
for vuz in VUZ_list:
    td = vuz.find('td')
    if td is None:
        continue
    
    a = td.find('a')
    if a is not None:
        l = a.get('href').split('/')[0]
        if l in ['misis', 'msu','mai']:
            continue
        links.append(a.get('href').split('/')[0])
```

```python
links_to_lists = []
for link in links:
    with urllib.request.urlopen(admlist_adress + '/' + link + '/index.html') as file:
        text = file.read().decode('utf-8')

    soup = BeautifulSoup(text, 'lxml')
    fac_list = soup.findAll('tr')

    for fac in fac_list:
        td = fac.find('td')
        if td is None:
            continue

        a = td.find('a')
        if a is not None:
            links_to_lists.append(admlist_adress + '/' + link + '/'+ a.get('href'))
print('Done')
```

```python
friends_found = []
for link in links_to_lists:    
    with urllib.request.urlopen(link) as file:
        text = file.read().decode('utf-8')

    soup = BeautifulSoup(text, 'lxml')
    print('Searching', soup.find('h2').text, '('+str(links_to_lists.index(link))+'/'+str(len(links_to_lists))+')')
    
    pupils_list = soup.findAll('tr')
    if len(pupils_list) <= 3:
        continue
    title = pupils_list[2]
    pupils_list = pupils_list[3:]
    
    for pupil in pupils_list:
        splits = pupil.findAll('td')[3].text.split()
        if len(splits) < 2:
            continue
        surname, name, *_ = splits
        name = surname + ' ' + name
        if name in friends:
            print('Found',name)
            friends_found.append(pupil)
            
print('Done')
```
