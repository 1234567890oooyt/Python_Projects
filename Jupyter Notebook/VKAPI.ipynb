```python
import json
import requests
members = json.loads(requests.get("https://api.vk.com/method/groups.getMembers?group_id=62351504").text) 

# Оставь надежду всяк сюда входящий!

#access_token = "0db277d901b53fc5e44bafc00840a3cd47af0fbfda8bd673992692a7e11753dc58f551b63aacf4cbeba01"
#json.loads(requests.get("https://api.vk.com/method/friends.get?user_id="+str(112634373)+"&count=5000&v=5.60&fields=id&access_token="+access_token).text)
```

```python
users = dict()
i = 0    
for id in members["response"]["users"]:
    users[id] = dict()
    users[id]["count"] = 0
    data = json.loads(requests.get("https://api.vk.com/method/users.get?user_id=" + str(id)).text)["response"][0]
    users[id]["name"] = data["first_name"] + " " + data["last_name"]
    users[id]["friendsID"] = json.loads(requests.get("https://api.vk.com/method/friends.get?user_id=" + str(id)).text)
    i += 1
    if i % 100 == 0: print(i)
```

```python

for user in users.keys(): 
    users[user]["count"] = 0
    try:
        for friendID in users[user]["friendsID"]["response"]:
            i = 0
            for f in users.keys():
                if f == friendID:
                    users[f]["count"] += 1
                if i < 0: 
                    i += 1
                    print(str(f), str(friendID))
    except:
        print("failed")
        pass
print("done")
```

```python
users2 = []
for userID in users.keys():
    users2.append({"name" : users[userID]["name"], "count" : users[userID]["count"]})
users2 = sorted(users2, key=lambda x: x["count"], reverse=True)

for user in users2:
    print(user)

file = open("D:\Python\Jupyter Notebook\popularity.csv", 'w', encoding='utf-8')
file.write("Name"+";"+ "Friends count\n")
for user in users2:
    file.write(str(user["name"] + ";" + str(user["count"]) + '\n'))
```
